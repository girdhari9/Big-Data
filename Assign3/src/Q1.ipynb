```python
import os
import requests
import pprint
import matplotlib.pyplot as plt
import requests
import operator
import numpy as np
from datetime import datetime
from dateutil import tz
MAC_URL = 'http://macvendors.co/api/%s'
baseUlr = "../dataset/Isis Admin's Hacked Phones/"
```

# Part 1:

```python
def fetchcompany(lines, mydict):
    for x in lines:
        try:
            if(x[:12] == 'Hardware MAC'):
                mac = x[15:-1]
                x = requests.get(MAC_URL % mac).json()
                c_name = x['result']['company']
                if(c_name in mydict):
                    mydict[c_name] += 1
                else:
                    mydict[c_name] = 1
        except:
            continue
```

```python
mydict, files = {}, os.listdir(baseUlr)
for name in files:
    filename = baseUlr + name + "/IS/ifconfig.txt"
    try:
        fetchcompany(tuple(open(filename, 'r')))
    except:
        f = 1
   
    for i in range(8):
        filename = baseUlr + name + "/IS" + str(i) + "/ifconfig"
        try:
            lines = tuple(open(filename, 'r'))
        except:
            filename = baseUlr + name + "/IS" + str(i) + "/ifconfig.txt"
            try:
                lines = tuple(open(filename, 'r'))
            except:
                continue
        fetchcompany(lines, mydict)
```

```python
plt.figure(figsize=(13, 6))
plt.ylabel('Company Name', fontsize=16)
plt.xlabel('Frequency of ISIS operatives', fontsize=14)
plt.yticks(range(len(mydict)), list(mydict.keys()))
plt.barh(range(len(mydict)),list(mydict.values()), align='center')
plt.show()
```

The Company whose phones is most popular among the ISIS operatives among all countries is XEROX CORPORATION

```python
count_num, count_country, count_name = {}, {}, {}
def fetch_calls(lines, callType, mycountry):
    for x in lines:
        s=x[-9:15]
        num=(x[:6])
        if(num=='Number'):
            prev_num=(x[9:])
        if(x[:4]=="Name"):
            prev_name=(x[7:])
        if(s == callType):
                if(prev_num.lower()[0]>='a' and  prev_num.lower()[0]<='z'):
                        continue
                if(prev_num in count_num):
                    count_num[prev_num]+=1
                else:
                    count_country[prev_num]=mycountry
                    count_name[prev_num]=prev_name
                    count_num[prev_num]=1
```

```python
def fetchCallDetail(callType):
    files = os.listdir(baseUlr)
    for name in files:
        filename = baseUlr + name + "/IS/Calllog Dump.txt"
        try:
            lines = tuple(open(filename, 'r'))
            fetch_calls(lines, callType, name)
        except:
            f=1

        for i in range(8):
            filename = baseUlr + name + "/IS" + str(i) + "/Calllog Dump"
            try:
                lines = tuple(open(filename, 'r'))
            except:
                filename = baseUlr + name + "/IS" + str(i) + "/Calllog Dump.txt"
                try:
                    lines = tuple(open(filename, 'r'))
                except:
                    continue
            fetch_calls(lines, callType, name)
```

```python
def localTime(time):
    Normal_time = tz.tzutc()
    Local_time = tz.tzlocal()
    utc = datetime.utcnow()
    utc = datetime.strptime(time, '%H:%M:%S')
    utc = utc.replace(tzinfo=Normal_time)
    IndiaTime = utc.astimezone(Local_time)
    return IndiaTime.time()
```

```python
callCount = {}
def fetch_call_detail(lines, mycountry):
    for x in lines:
        date = (x[:4])
        localtime = ""
        if(date == 'Date'):
            x = x.split(" ")
            if len(x[2]) == 3:
                localtime = localTime(x[4])
            else:
                localtime = localTime(x[2])
            hour = int(str(localtime).split(':')[0])
            if mycountry in callCount:
                callCount[mycountry][int(hour-1)] += 1
            else:
                callCount[mycountry] = [0 for i in range(24)]
```

```python
files = os.listdir(baseUlr)
country = ['India IS (Admins)', 'Syria IS (Admins)', 'Yemen IS (Admins)']
for name in files:
    if name in country:
        filename = baseUlr + name + "/IS/Calllog Dump.txt"
        try:
            lines = tuple(open(filename, 'r'))
            fetch_call_detail(lines, name.split(" ")[0])
        except:
            f=1

        for i in range(8):
            filename = baseUlr + name + "/IS" + str(i) + "/Calllog Dump"
            try:
                lines = tuple(open(filename, 'r'))
            except:
                filename = baseUlr + name + "/IS" + str(i) + "/Calllog Dump.txt"
                try:
                    lines = tuple(open(filename, 'r'))
                except:
                    continue
            fetch_call_detail(lines, name.split(" ")[0])
```
