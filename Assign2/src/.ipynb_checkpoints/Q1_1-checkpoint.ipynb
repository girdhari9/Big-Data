```python
import sys
import pandas as pd
import numpy as np
from numpy import *
import scipy.stats as stats
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
if __name__ == "__main__":
    df = pd.read_csv("../dataset/baltimore_crimes.csv")
    df = df['District']
    unique_values = np.unique(df)
```

```python
result = {}
for uv in unique_values:
    result[uv] = 0
```

```python
for i in range(len(df)):
    result[df[i]] += 1
```

```python
width = 1/5
barlist = plt.bar(unique_values, result.values(), width, color="red")
barlist[3].set_color('b')
barlist[2].set_color('g')
plt.xlabel('District')
plt.ylabel('Incident')
plt.title('Frequency of crime incidents across all districts')
plt.xticks(rotation='vertical')
plt.show()
```
