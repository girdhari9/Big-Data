```python
import sys
import pandas as pd
import numpy as np
from numpy import *
import scipy.stats as stats
import random
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
```

```python
if __name__ == "__main__":
    df = pd.read_csv("../dataset/global_terrorism_database.csv")
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(df['eventid'])
```

```python
edges = df[['related','eventid']].dropna()
for index in range(len(edges)):
    edge = edges.iloc[[index]]
    fr = int(edge['eventid'].values[0])
    to = edge['related'].values[0].split(',')
    for x in to:
        try:
            if int(x) != fr:
                G.add_edge(fr, int(x))
        except:
            continue
```

```python
G.number_of_edges()
```

```python
G.number_of_nodes()
```

```python
maxCompponent = list(max(nx.connected_components(G), key = len))
```

```python
newGraph = nx.Graph()
newGraph.add_nodes_from(maxCompponent)
for i in G.edges():
    if i[0] in maxCompponent and i[1] in maxCompponent:
        newGraph.add_edge(i[0],i[1])
```

```python
nx.draw(newGraph)
```
