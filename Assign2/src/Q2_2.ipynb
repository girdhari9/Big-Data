```python
import sys
import pandas as pd
import numpy as np
from numpy import *
import scipy.stats as stats
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
if __name__ == "__main__":
    df = pd.read_csv("../dataset/global_terrorism_database.csv")
```

```python
years = np.unique(df['iyear'])
```

```python
result = {}
for year in years:
    result[year] = 0
for index in range(1, len(df), 1):
    result[df.iloc[[index]]['iyear'].values[0]] += 1
```

```python
width = 1/5
plt.figure(num=None, figsize=(15, 6), dpi=70, facecolor='w', edgecolor='k')
barlist = plt.bar(years, result.values(), width, color="red")
barlist[-4].set_color('b')
plt.xlabel('Years')
plt.ylabel('Crimes Rate')
plt.title('Years vs Crimes Rate')
plt.xticks(rotation='vertical')
plt.show()
```

```python
maxCrime, maxCrimeYear = 0, 0
for res in result:
    if maxCrime < result[res]:
        maxCrime = result[res]
        maxCrimeYear = res
```

```python
countries = np.unique(df['region_txt'])
```

```python
newResult = {}
for country in countries:
    newResult[country] = 0
for index in range(1, len(df), 1):
    if df.iloc[[index]]['iyear'].values[0] == maxCrimeYear:
        newResult[df.iloc[[index]]['region_txt'].values[0]] += 1
```

```python
width = 1/5
plt.figure(num=None, figsize=(15, 6), dpi=70, facecolor='w', edgecolor='k')
barlist = plt.bar(newResult.keys(), newResult.values(), width, color="red")
barlist[5].set_color('b')
plt.xlabel('Continent')
plt.ylabel('Crime\'s Rate')
plt.title('Continent vs Crime\'s Rate in' + str(maxCrimeYear))
plt.xticks(rotation='vertical')
plt.show()
```
