```python
import sys
import pandas as pd
import numpy as np
from numpy import *
import scipy.stats as stats
import random
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS 
```

```python
if __name__ == "__main__":
    df = pd.read_csv("../dataset/global_terrorism_database.csv")
```

```python
stopwords = set(STOPWORDS)
stopwords.update(['Nan','Unknown','attack','specific','motive','source','sources','noted','started','responsibility','carried','incident'])
```

```python
summary = df['summary'].str.cat(sep = ' ')
```

```python
wordcloud = WordCloud(width = 800, height = 800, background_color ='white', stopwords = stopwords, min_font_size = 10).generate(summary) 
```

```python
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```

```python
motive = df['motive'].str.cat(sep = ' ')
```

```python
Motivewordcloud = WordCloud(width = 800, height = 800, background_color ='white', stopwords = stopwords, min_font_size = 10).generate(motive) 
```
