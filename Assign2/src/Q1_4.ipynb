```python
import sys
import pandas as pd
import numpy as np
from numpy import *
import scipy.stats as stats
import random
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import gmplot
from IPython.display import IFrame
```

```python
if __name__ == "__main__":
    df = pd.read_csv("../dataset/baltimore_crimes.csv")
```

```python
CrimeCodes = np.unique(df['CrimeCode'])
```

```python
result = {}
for index in range(1, len(df), 1):
    data = df.iloc[[index]]
    if data['Description'].values[0] not in result:
        result[data['Description'].values[0]] = 0
    result[data['Description'].values[0]] += 1
```

```python
width = 1/5
plt.figure(num=None, figsize=(15, 6), dpi=70, facecolor='w', edgecolor='k')
barlist = plt.bar(result.keys(), result.values(), width, color="red")
barlist[0].set_color('b')
plt.xlabel('Types of Crime')
plt.ylabel('Frequency')
plt.title('Types of Crime vs Frequency')
plt.xticks(rotation='vertical')
plt.show()
```

```python
mostFrequentCrime = 0
for key in result:
    if result[key] > mostFrequentCrime:
        mostFrequentCrime = result[key]
        code = key
```

```python
newResult = {}
for index in range(1, len(df), 1):
    data = df.iloc[[index]]
    if data['Description'].values[0] == code:
        time = list(map(int,data['CrimeTime'].values[0].split(':')))
        if time[0] <= 23 and time[0] >= 0 and time[0] not in newResult:
            newResult[time[0]] = 0
        newResult[time[0]] += 1
```

```python
width = 1/5
plt.figure(num=None, figsize=(15, 6), dpi=70, facecolor='w', edgecolor='k')
barlist = plt.bar(newResult.keys(), newResult.values(), width, color="red")
plt.xlabel('Types of Crime')
plt.ylabel('Frequency of Crimes')
plt.title('Types of Crime vs Frequency of Crimes of Most Frequent Crime')
plt.xticks(rotation='vertical')
plt.show()
```

```python
lat, long =  [], []
LatLong = df['Location 1']
CrimeCode = df['Description']
for index in range(len(LatLong)):
    if CrimeCode[index] == code:
        loc = LatLong[index].split(',')
        lat.append(float(loc[0][1:]))
        long.append(float(loc[1][1:-1]))
```

```python
gmap = gmplot.GoogleMapPlotter(np.mean(lat), np.mean(long), 13)
gmap.heatmap(lat, long)
gmap.draw("mymap.html")
IFrame("mymap.html", width = 1000, height = 400)
```
